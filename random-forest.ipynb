```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
iris
```

```python
import pandas as pd
data = pd.DataFrame({"Sepal length": iris.data[:,0],
                    "Sepal width": iris.data[:,1],
                   "Petal length": iris.data[:,2],
                    "Petal width": iris.data[:,3],
                    "species": iris.target})
```

```python
data
```

x_train are the features of the train dataset, x_test are the features of the test dataset
y_train are the output labels of the train dataset, y_test are the output labels of the test dataset

```python
#model training : train test splitting
from sklearn.model_selection import train_test_split
x=data[['Sepal length','Sepal width','Petal length','Petal width']]
y=data['species']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20, random_state = 42)
```

```python
#model definition
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.predict(x_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_pred, y_test)
```
